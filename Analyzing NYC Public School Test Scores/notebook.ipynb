# Reading Data + EDA

```python
schools_df = pd.read_csv('schools.csv')
schools_df.head()
```

```python
schools_df.info()
```

```python
schools_df.describe()
```

## Schools with 80th percentile and above math scores

```python
# Find the best schools for math
best_math_schools = schools_df[schools_df["average_math"] >= 640][["school_name", "average_math"]].sort_values("average_math", ascending=False)
best_math_schools
```

## Top 10 Schools with best SAT scores

```python
top_10_schools = schools_df
top_10_schools['average_SAT'] = top_10_schools['average_math'] + top_10_schools['average_reading'] + top_10_schools['average_writing']
top_10_schools['average_SAT']
```

```python
top_10_schools = top_10_schools.loc[:, ['school_name', 'average_SAT']].sort_values(by="average_SAT", ascending=False).head(10)
top_10_schools
```

## Borough with highest SAT score standard deviation

```python
largest_std_dev = schools_df
largest_std_dev['average_SAT'] = largest_std_dev['average_math'] + largest_std_dev['average_reading'] + largest_std_dev['average_writing']
```

```python
boroughs = schools_df.groupby("borough")["average_SAT"].agg(["count", "mean", "std"]).round(0).sort_values("mean", ascending=False)
```

```python
largest_std_dev = boroughs[boroughs["std"] == boroughs["std"].max()]
largest_std_dev
```

```python
# Share your results in this format
schools_analysis = {"best_math_schools": best_math_schools,
                    "top_10_schools": top_10_schools,
                    "largest_SAT_std_dev": largest_std_dev}

# Call the answer!
print(schools_analysis)
```
